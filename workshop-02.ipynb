```python
import psutil, time, csv
from datetime import datetime
import pathlib

def write_to_csv(writer, i, row):
    print(f"{i}: {row}")
    writer.writerow(row)


fname = r"C:\Users\choon\Documents\pydot23\04-Day04\cpu_and_memory.csv"
if not pathlib.Path(fname).exists():    
    with open(fname, mode='w', newline='') as f:
        csv_writer = csv.writer(f)
        row = ["cpu_percent", "memory_percent", "current_time"]
        write_to_csv(csv_writer, 0, row)


with open(fname, mode='a', newline='') as f:
    csv_writer = csv.writer(f)

    duration = 20

    for i in range(duration):
        row = [psutil.cpu_percent(), psutil.virtual_memory().percent, datetime.now()]
        write_to_csv(csv_writer, i + 1, row)

        time.sleep(1)
        i += 1
```
